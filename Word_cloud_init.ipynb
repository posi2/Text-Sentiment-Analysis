```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
from textblob import TextBlob
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/posi2/Text-Sentiment-Analysis/master/final_review.csv",header=None,names=('text','sentiment'))
```

```python
data.head()
```

```python
df_positive=data.loc[data['sentiment']==1]
df_positive.reset_index(inplace=True)
df_neutral=data.loc[data['sentiment']==0]
df_neutral.reset_index(inplace=True)
df_negative=data.loc[data['sentiment']==-1]
df_negative.reset_index(inplace=True)
```

```python
stopwords = set(STOPWORDS)
```

```python
full_text= data['text'].astype(str).apply(lambda x: ', '.join(data['text'].astype(str)))
full_text=full_text[2486]

positive_text= df_positive['text'].astype(str).apply(lambda x: ', '.join(df_positive['text'].astype(str)))
positive_text=positive_text[828]

neutral_text= df_neutral['text'].astype(str).apply(lambda x: ', '.join(df_neutral['text'].astype(str)))
neutral_text=neutral_text[828]

negative_text= df_negative['text'].astype(str).apply(lambda x: ', '.join(df_negative['text'].astype(str)))
negative_text=negative_text[828]
```

```python
# instantiate a word cloud object
word_cloud = WordCloud(
    background_color='white',
    max_words=2000000,
    stopwords=stopwords
)
```

## Full-Text Wordcloud

```python
# generate the word cloud
word_cloud.generate(full_text)

fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## Positive-Text Wordcloud

```python
word_cloud.generate(positive_text)

fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## Neutral-Text Wordcloud

```python
word_cloud.generate(neutral_text)

fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
word_cloud.generate(negative_text)

fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
